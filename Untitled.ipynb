```python
import pandas as pd

# Ruta al archivo CSV original
csv_path = "train.csv"

# Número de filas por archivo dividido
chunk_size = 100_000

# Usamos un iterador para no cargar todo en memoria
reader = pd.read_csv(csv_path, chunksize=chunk_size)

# Guardamos cada chunk como un nuevo archivo CSV
for i, chunk in enumerate(reader):
    output_path = f"train_part_{i+1}.csv"
    chunk.to_csv(output_path, index=False)
    print(f"✔ Guardado: {output_path}")
```
